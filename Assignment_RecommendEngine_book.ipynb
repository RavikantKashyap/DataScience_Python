```python
#User Based Collaborative Filtering
import pandas as pd
import numpy as np
```

```python
book=pd.read_csv("E:/Ravi_ExcelR_Assignments/Recommendation system/book.csv",encoding = "ISO-8859-1")
book
```

```python
book.columns
```

```python
book =book.rename(columns={"User.ID":"User_ID","Book.Title":"Book_Title","Book.Rating":"Book_Rating"})
```

```python
#number of unique users in the dataset
len(book.User_ID.unique())
```

```python
#number of unique users in the dataset
len(book.Book_Title.unique())
```

```python
book.drop_duplicates(subset=None, keep='first', inplace=True)
```

```python
book=book.drop_duplicates(subset=["Book_Title"])
```

```python
book.shape
```

```python
book_df = book.pivot(index='User_ID',
                                 columns='Book_Title',
                                 values='Book_Rating').reset_index(drop=True)
```

```python
book_df
```

```python
book_df.index = book.User_ID.unique()
```

```python
book_df
```

```python
#Impute those NaNs with 0 values
book_df.fillna(0, inplace=True) 
```

```python
book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book_df.values,metric='cosine') 
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book.User_ID.unique()
user_sim_df.columns = book.User_ID.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```

```python
Decision_Normandy_rating = book_df['Decision in Normandy']  
```

```python
#Finding the correlation with different movies
similar_to_Normandy = book_df.corrwith(Decision_Normandy_rating) 
```

```python
similar_to_Normandy.head(10) 
```

```python
corr_Normanding = pd.DataFrame(similar_to_Normandy, columns=['Correlation'])
corr_Normanding.dropna(inplace=True)
corr_Normanding.head()
```

```python
corr_Normanding[corr_Normanding['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)  
```

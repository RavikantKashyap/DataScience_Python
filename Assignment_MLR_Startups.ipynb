```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
```

```python
startup=pd.read_csv("E:/Ravi_ExcelR_Assignments/Multiple Linear Regression/50_Startups.csv")
startup
```

```python
#filling categorical variable with dummy values
startup_data=pd.get_dummies(startup,columns=['State'])  
```

```python
startup_data
```

```python
#checking null values
startup_data.info()
```

```python
#check for missing values
startup_data.isna().sum()
```

```python
startup_data.corr()
```

```python
startup_data=pd.DataFrame(startup_data)
```

```python
#startup=pd.DataFrame(startup)
```

```python
startup_data.rename(columns = {'State_New York':'State_New_York'}, inplace = True) 
```

```python
startup_data.rename(columns = {'R&D_Spend':'RD_Spend'}, inplace = True) 
```

```python
#Build model
import statsmodels.formula.api as smf 
model = smf.ols('Profit~Administration+Marketing_Spend+State_California+State_New_York+RD_Spend+State_Florida',data=startup_data).fit()
```

```python
#Coefficients
model.params
```

```python
model.summary()
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

```python
#administratiion and marketing spend p values are >0.05 so insignificant
```

```python
#Profit on administration
ml_adm=smf.ols('Profit~Administration',data = startup_data).fit()  
#t and p-Values
print(ml_adm.tvalues, '\n', ml_adm.pvalues) 
```

```python
#Profit on marketing spend
ml_mkt=smf.ols('Profit~Marketing_Spend',data = startup_data).fit()  
#t and p-Values
print(ml_mkt.tvalues, '\n', ml_mkt.pvalues) 
```

```python
#individually marketing spend p value is <0.05 but administration p value is >0.05 
```

```python
#Together
ml_mktadm=smf.ols('Profit~Marketing_Spend+Administration',data = startup_data).fit()  
#t and p-Values
print(ml_mktadm.tvalues, '\n', ml_mktadm.pvalues) 
```

```python
#together pvalues<0.05 its significant but individually Administration is insignificant as pvalue>0.05
```

```python
#Checking the Varaince Inflation Factor
```

```python
rsq_adm=smf.ols("Administration~Marketing_Spend+RD_Spend+State_California+State_New_York+State_Florida",data=startup_data).fit().rsquared
vif_adm=1/(1-rsq_adm)
vif_adm
```

```python
rsq_mkt=smf.ols("Marketing_Spend~Administration+RD_Spend+State_California+State_New_York+State_Florida",data=startup_data).fit().rsquared
vif_mkt=1/(1-rsq_mkt)
vif_mkt
```

```python
rsq_rd=smf.ols("RD_Spend~Administration+Marketing_Spend+State_California+State_New_York+State_Florida",data=startup_data).fit().rsquared
vif_rd=1/(1-rsq_rd)
vif_rd 
```

```python
#VIF high for RD_Spend
#Doubts: VIF infinity for all 3 states , so need to consider or not
```

```python
#cooks distance calculation
model_influence=model.get_influence()
(c,_)=model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(startup_data)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# observations lies <1 so no influencing observations
```

```python
#Exclude variable "Administration" and generate R-Squared and AIC values
startup_data_adm= smf.ols('Profit~Marketing_Spend+RD_Spend+State_California+State_New_York+State_Florida',data = startup_data).fit()
```

```python
(startup_data_adm.rsquared,startup_data_adm.aic)
```

```python
#Exclude variable "RD_Spend" and generate R-Squared and AIC values
startup_data_rds= smf.ols('Profit~Marketing_Spend+Administration+State_California+State_New_York+State_Florida',data = startup_data).fit()
```

```python
(startup_data_rds.rsquared,startup_data_rds.aic)
```

```python
#as Rd_Spend aic is high and rsquared is low compare to administration, so Rd_Spend can be removed 
#now check influencing observations using cooks distance
```

```python
model_influence_adm=startup_data_adm.get_influence()
(c_adm,_)=model_influence_adm.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(startup_data)), np.round(c_adm, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#no influencing observations as observations lies <1 so we can stop here
```

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt 
```

```python
crime_data = pd.read_csv("E:/Ravi_ExcelR_Assignments/Clustering/crime_data.csv")
crime_data
```

```python
#normalizing the data
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(crime_data.iloc[:,1:])
df_norm
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
z = linkage(df_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
crime_data['clust']=cluster_labels # creating a  new column and assigning it to new column 
crime_data
```

```python
crime_data.iloc[:,1:].groupby(crime_data.clust).mean()
```

```python
crime_data = pd.read_csv("E:/Ravi_ExcelR_Assignments/Clustering/crime_data.csv")
crime_data
```

```python
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x)
```

```python
df_norm = norm_func(crime_data.iloc[:,1:])
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 20):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 20), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(df_norm)
```

```python
y_kmeans
#clf.cluster_centers_
clf.labels_
```

```python
md=pd.Series(y_kmeans)  # converting numpy array into pandas series object 
crime_data['clust']=md # creating a  new column and assigning it to new column 
crime_data
```

```python
crime_data.iloc[:,1:5].groupby(crime_data.clust).mean()
```

```python
crime_data.plot(x="Rape",y ="Murder",c=clf.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```
